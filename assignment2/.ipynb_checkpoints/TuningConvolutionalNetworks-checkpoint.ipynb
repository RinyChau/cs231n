## Load Data and packages

```python
# As usual, a bit of setup

import numpy as np
import matplotlib.pyplot as plt
from cs231n.classifiers.cnn import *
from cs231n.data_utils import get_CIFAR10_data
from cs231n.gradient_check import eval_numerical_gradient_array, eval_numerical_gradient
from cs231n.layers import *
from cs231n.fast_layers import *
from cs231n.solver import Solver
import timeit

%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'

# for auto-reloading external modules
# see http://stackoverflow.com/questions/1907993/autoreload-of-modules-in-ipython
%load_ext autoreload
%autoreload 2

def rel_error(x, y):
  """ returns relative error """
  return np.max(np.abs(x - y) / (np.maximum(1e-8, np.abs(x) + np.abs(y))))
```

```python
# Load the (preprocessed) CIFAR10 data.

data = get_CIFAR10_data()
for k, v in data.iteritems():
  print '%s: ' % k, v.shape
```

```python
def train(layer):
    kwargs = {}
    learning_rate = 1e-3
    if 'reg' in layer[0]['params']:
        kwargs['reg'] = layer[0]['params']['reg']
        del layer[0]['params']['reg']
    if 'learning_rate' in layer[0]['params']:
        learning_rate = layer[0]['params']['learning_rate']
        del layer[0]['params']['learning_rate']
    model = FlexibleConvNet(layers=layer, dtype=np.float64, use_xavier=True,**kwargs)
    solver = Solver(model, data, num_epochs=num_epochs, batch_size=500, update_rule='adam',
                        optim_config={
                            'learning_rate': learning_rate
                        },
                        verbose=False)
    solver.model.name = modelName(solver)
    solver.train()
    return {"name": solver.model.name, "val_acc_history": solver.val_acc_history, "loss_history":solver.loss_history,
          "train_acc_history":solver.train_acc_history }
    
```

```python
def plot_train_history(results):
    plt.subplot(3, 1, 1)
    plt.title('Training loss')
    plt.xlabel('Iteration')

    plt.subplot(3, 1, 2)
    plt.title('Training accuracy')
    plt.xlabel('Epoch')

    plt.subplot(3, 1, 3)
    plt.title('Validation accuracy')
    plt.xlabel('Epoch')
    
    for result in results:
        plt.subplot(3, 1, 1)
        plt.plot(result['loss_history'], 'o', label=result['name'])
        plt.subplot(3, 1, 2)
        
        plt.plot(result['train_acc_history'], '-o', label=result['name'])
        plt.subplot(3, 1, 3)
        plt.plot(result['val_acc_history'], '-o', label=result['name'])
  
    for i in [1, 2, 3]:
        plt.subplot(3, 1, i)
        plt.legend(loc='upper center', ncol=4)
    plt.gcf().set_size_inches(15, 15)
    plt.show()
```

# Different Reg and Learning Rate

```python
import math
layers = []
solvers = []
modelName = lambda x: str(x.model.reg) + ":"+ str(x.optim_config['learning_rate'])
num_epochs=20
regularization_strengths =[1e-4, 3e-4, 1e-3, 3e-3, 1e-2 , 2e-2, 0]
learning_rates = np.logspace(math.log(1e-4,10), math.log(1e-2,10), num=5)
for reg in regularization_strengths:
    for learning_rate in learning_rates:
        layer = []
        layer.append({'func':'conv_spatial_batchnorm_relu_pool','params':{'num_filters':64, 'filter_size':3, 'reg':reg, 
                                                                          'learning_rate':learning_rate}})
        layer.append({'func':'conv_spatial_batchnorm_relu_pool','params':{'num_filters':128, 'filter_size':3}})
        layer.append({'func':'conv_spatial_batchnorm_relu_pool','params':{'num_filters':256, 'filter_size':3}})
        layer.append({'func':'affine_batch_norm_relu','params':{'hidden_dim':400}})
        layer.append({'func':'affine_batch_norm_relu','params':{'hidden_dim':400,}})
        layer.append({'func':'affine_batch_norm_relu','params':{'hidden_dim':100}})
        layers.append(layer)

from multiprocessing import Pool
p = Pool(len(layers))

start = timeit.default_timer()
results = p.map(train, layers)
stop = timeit.default_timer()
plot_train_history(results)
print stop - start 
```
